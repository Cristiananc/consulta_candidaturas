```python
df_vagas_2014 = pd.read_csv('relacao_vaga_2018.csv', sep=';', encoding='latin1', low_memory=False, usecols=['UF','Cargo', 'Quantitativo', 'Vaga'])

df_deputados_estaduais_2014 = df_vagas_2014[df_vagas_2014['Cargo'] == 'Deputado Estadual']

df_deputados_estaduais_2014 = df_deputados_estaduais_2014.rename(columns={'UF': 'SG_UF', 'Vaga': 'VAGAS'});
```

```python
df_deputados_estaduais_2014
```

### Candidatos a deputados estaduais em 2014

Abaixo estão as coligações presentes em cada estado e a quantidade de candidaturas por gênero em cada coligação, além da quantidade de candidaturas por gênero representadas por partidos isolados:

### Deputados estaduais com coligação 

Respostas para o campo de 'DS_GENERO':

```python
df_2014['DS_GENERO'].unique()
```

Gênero Feminino

```python
df_deputado_estadual_com_coligacao_genero_feminino_2014 = df_2014[
    (df_2014['DS_CARGO'] == 'DEPUTADO ESTADUAL') &
    (df_2014['DS_GENERO'] == 'FEMININO') &
    (df_2014['TP_AGREMIACAO'] == 'COLIGAÇÃO') &
    (df_2014['DS_SITUACAO_CANDIDATURA'] == 'APTO')
].groupby(['SG_UF','NM_UE', 'SQ_COLIGACAO', 'NM_COLIGACAO'], as_index = False)['DS_GENERO'].count()
```

```python
df_deputado_estadual_com_coligacao_genero_feminino_2014 = df_deputado_estadual_com_coligacao_genero_feminino_2014.rename(columns={'DS_GENERO':'GENERO_FEMININO'});
```

```python
df_deputado_estadual_com_coligacao_genero_feminino_2014.head()
```

Gênero Masculino

```python
df_deputado_estadual_com_coligacao_genero_masculino_2014 = df_2014[
    (df_2014['DS_CARGO'] == 'DEPUTADO ESTADUAL') &
    (df_2014['DS_GENERO'] == 'MASCULINO') &
    (df_2014['TP_AGREMIACAO'] == 'COLIGAÇÃO') &
    (df_2014['DS_SITUACAO_CANDIDATURA'] == 'APTO')
].groupby(['SG_UF','NM_UE', 'SQ_COLIGACAO', 'NM_COLIGACAO'], as_index = False)['DS_GENERO'].count()
```

```python
df_deputado_estadual_com_coligacao_genero_masculino_2014 = df_deputado_estadual_com_coligacao_genero_masculino_2014.rename(columns={'DS_GENERO':'GENERO_MASCULINO'});
```

```python
df_deputado_estadual_com_coligacao_genero_masculino_2014.head()
```

Genêro não divulgável

```python
df_deputado_estadual_com_coligacao_nao_divulgavel_2014 = df_2014[
    (df_2014['DS_CARGO'] == 'DEPUTADO ESTADUAL') &
    (df_2014['DS_GENERO'] == 'NÃO DIVULGÁVEL') &
    (df_2014['TP_AGREMIACAO'] == 'COLIGAÇÃO') &
    (df_2014['DS_SITUACAO_CANDIDATURA'] == 'APTO')
].groupby(['SG_UF','NM_UE', 'SQ_COLIGACAO', 'NM_COLIGACAO'], as_index = False)['DS_GENERO'].count()
df_deputado_estadual_com_coligacao_nao_divulgavel_2014 = df_deputado_estadual_com_coligacao_nao_divulgavel_2014.rename(columns={'DS_GENERO':'GENERO_NAO_DIV'});
```

```python
df_deputado_estadual_com_coligacao_nao_divulgavel_2014.head()
```

Colocando os resultados numa mesma tabela:

```python
df_deputado_estadual_com_coligacao_por_genero_2014 = pd.merge(pd.merge(df_deputado_estadual_com_coligacao_genero_feminino_2014, df_deputado_estadual_com_coligacao_genero_masculino_2014[['SQ_COLIGACAO', 'GENERO_MASCULINO']], on = 'SQ_COLIGACAO', how = 'outer'), df_deputado_estadual_com_coligacao_nao_divulgavel_2014[['SQ_COLIGACAO', 'GENERO_NAO_DIV']], on = 'SQ_COLIGACAO', how = 'outer') 
df_deputado_estadual_com_coligacao_por_genero_2014 = df_deputado_estadual_com_coligacao_por_genero_2014.fillna(0)
```

```python
df_deputado_estadual_com_coligacao_por_genero_2014.head()
```

Total de candidatos por coligação por município:

```python
df_deputado_estadual_com_coligacao_por_genero_2014['TOTAL'] = df_deputado_estadual_com_coligacao_por_genero_2014['GENERO_FEMININO'] + df_deputado_estadual_com_coligacao_por_genero_2014['GENERO_MASCULINO'] + df_deputado_estadual_com_coligacao_por_genero_2014['GENERO_NAO_DIV'] 
```

Porcentagem de candidatos por gênero

```python
df_deputado_estadual_com_coligacao_por_genero_2014['PERC_FEM'] = (df_deputado_estadual_com_coligacao_por_genero_2014['GENERO_FEMININO']/df_deputado_estadual_com_coligacao_por_genero_2014['TOTAL'])*100
df_deputado_estadual_com_coligacao_por_genero_2014['PERC_MASC'] = (df_deputado_estadual_com_coligacao_por_genero_2014['GENERO_MASCULINO']/df_deputado_estadual_com_coligacao_por_genero_2014['TOTAL'])*100
df_deputado_estadual_com_coligacao_por_genero_2014['PERC_NAO_DIV'] = (df_deputado_estadual_com_coligacao_por_genero_2014['GENERO_NAO_DIV']/df_deputado_estadual_com_coligacao_por_genero_2014['TOTAL'])*100
```

```python
df_deputado_estadual_com_coligacao_por_genero_2014 = pd.merge(df_deputado_estadual_com_coligacao_por_genero_2014, df_deputados_estaduais_2014[['SG_UF', 'VAGAS']] , on = 'SG_UF')
```

```python
df_deputado_estadual_com_coligacao_por_genero_2014['PERCENT_OCUPADO'] = (df_deputado_estadual_com_coligacao_por_genero_2014['TOTAL']/df_deputado_estadual_com_coligacao_por_genero_2014['VAGAS'])*100
```

```python
df_deputado_estadual_com_coligacao_por_genero_2014.head()
```
